#### (a) Take a grayscale image of size 512x512, perform the brightness enhancement of a specific range of gray levels & observe its result

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Loading the original image


```python
original_image = cv2.imread('./images/Fractured_Spine.tif',0)

print(f'shape of the image: {original_image.shape}')
plt.imshow(original_image,cmap='gray')
plt.title('Original Image')
plt.show()
```

### Function for Enhancing Brightness 

```python
def BrightnessEnhancement(original_image):
    height, width = original_image.shape
    for i in range(height):
        for j in range(width):
            if original_image[i,j]>=10 and original_image[i,j]<=50:
                original_image[i,j] = original_image[i,j] + 50 #50 is extra-brightness
            original_image[i,j] = min(255,original_image[i,j]) 
            
    return original_image         

         
        
```

### Enhancing Brightness

```python
BrightNessimage = BrightnessEnhancement(original_image)
plt.imshow(BrightNessimage,cmap='gray')
plt.show()
```
