```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
frequencies = np.linspace(0,100,500)
cutoff_frequency = 30
order = 4

butterWorth_filter_response = 1/(1+(frequencies/cutoff_frequency)**(2*order))
gaussian_filter_response = np.exp(-frequencies**2 / (2*cutoff_frequency**2))

plt.figure(figsize=(10, 6))
plt.plot(frequencies, butterWorth_filter_response, label="Butterworth Filter (4th Order)", linewidth=2)
plt.plot(frequencies, gaussian_filter_response, label="Gaussian Filter", linewidth=2, linestyle='--')
plt.axvline(cutoff_frequency, color='red', linestyle=':', label="Cutoff Frequency (D0 = 30)")

plt.title("Frequency Response of Butterworth and Gaussian Low-Pass Filters", fontsize=14)
plt.xlabel("Frequency", fontsize=12)
plt.ylabel("Filter Response (H(u, v))", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
