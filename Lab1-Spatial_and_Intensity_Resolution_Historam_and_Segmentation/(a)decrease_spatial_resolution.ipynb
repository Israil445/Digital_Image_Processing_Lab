####  (a) Take a grayscale image of size 512x512, decrease its spatial resolution by half every time & observe it's change when displaying in the same window size

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Loading the orignal image

```python
original_image = cv2.imread('./images/flower.jpg')

print(f'shape of the image: {original_image.shape}')


plt.imshow(original_image)
plt.axis('off')
plt.title('BGR format')
plt.show()
```

### resize and then convert it into RGB format

```python
#resizing the image
original_image = cv2.resize(original_image,(512,512))
print(f"shape after resize: {original_image.shape}")

#converting RGB 
rgb_image = cv2.cvtColor(original_image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.axis('off')
plt.title('RGB format')
plt.show()
```

### converting RGB to GRAY


```python
#converting RGB 
gray_image = cv2.cvtColor(original_image,cv2.COLOR_RGB2GRAY)
plt.imshow(rgb_image,cmap='gray')
plt.axis('off')
plt.title('Gray format')
plt.show()
```

### function for decreasing spatial resolution by half

```python
def decrease_resolution(image):
    height,width = image.shape
    decrease_image = np.zeros((height//2, width//2)) #new 2D array
    
    for i in range(0,height,2):
        for j in range(0,width,2):
            decrease_image[i//2,j//2] = image[i,j]
            
    return decrease_image        
            
```

### decreasing tthe spatial resolution by half

```python
decrease_image = gray_image
plt.figure(figsize=(8,10))

for i in range(1,5):
    plt.subplot(2,2,i)
    plt.imshow(decrease_image,cmap='gray')
    height, width = decrease_image.shape
    plt.title(f"{height}x{width}")
    decrease_image = decrease_resolution(decrease_image)
    
plt.show()    
```
