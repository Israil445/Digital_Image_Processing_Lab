#### (b) Take a grayscale image of size 512x512, decrease its intensity level resolution by one bit up to reach its binary format & observe its change when displaying in the same window size

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

### Loading the original image

```python
original_image = cv2.imread('./images/Skull.tif',0)


print(f'shape of the image: {original_image.shape}')


plt.imshow(original_image,cmap='gray')
plt.axis('off')
plt.title('8-bit image')
plt.show()

#convet the image into numpy array
arr = np.array(original_image)
# Display the array, max, and min values
print(f"Max Pixel Value: {np.max(arr)}")
print(f"Min Pixel Value: {np.min(arr)}")
```

### function for decreasing intensity level resolution 

```python
def decrease_intensity_resolution(image,bits):
    step = 255/(2**bits - 1)
    height,width = image.shape
    decrease_image = image.copy()
    
    for i in range(height):
        for j in range(width):
            decrease_image[i,j] = round(image[i,j]/step) * step
            
    return decrease_image      
```

### decreasing intensity level resolution by 1-bit each time

```python
decrease_image = original_image.copy()
plt.figure(figsize=(13,8))

for i in range(1,9):
    plt.subplot(2,4,i)
    bits = 9 - i
    decrease_image = decrease_intensity_resolution(decrease_image,bits)
    plt.imshow(decrease_image,cmap='gray')
    plt.title(f'{bits}-bits image')
    
plt.tight_layout()    
plt.show()
```
