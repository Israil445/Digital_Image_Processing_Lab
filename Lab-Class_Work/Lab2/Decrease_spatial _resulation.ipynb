```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#load the image
# Read image using Matplotlib
image_bgr = cv2.imread('rose.jpg')
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
image_rgb.shape
```

```python
# Resize the image
original_image = cv2.resize(image_rgb, (512, 512))

# Display the resized image
plt.imshow(original_image)
plt.axis('off')  # Hide axis
plt.show()
print(original_image.shape)
```

```python
#convert the image into gray scale
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='gray')
```

```python
gray_image.shape
```

```python
#decrease the spatial resulation 
def decrease_spatial_resulation(image):
    height,width = image.shape
    # Calculate new dimensions (half the original size)
    new_width = width // 2
    new_height = height // 2

   # Resize the image to half its original resolution
    resized_image = cv2.resize(image, (new_width, new_height))
    
    return resized_image
    
```

```python

plt.figure(figsize = (13, 13))
decreased_image = np.copy(gray_image)
for k in range (1,7):
    plt.subplot(2, 3, k)
    plt.imshow(decreased_image, cmap = 'gray')
    height, width = decreased_image.shape
    plt.title(f"{height}x{width}")
    decreased_image = decrease_spatial_resulation(decreased_image)

plt.show()
```
