```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#load the image
# Read image using Matplotlib
image_bgr = cv2.imread('rose.jpg')
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
# Resize the image
original_image = cv2.resize(image_rgb, (512, 512))

# Display the resized image
plt.imshow(original_image)
plt.axis('off')  # Hide axis
plt.show()
print(original_image.shape)
```

```python
#convert the image into gray scale
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='gray')
```

```python
def Decrease_Intensity_Resolution(image,bits):
    mxIntensity = 2**bits - 1
    return np.floor(image/(256//(mxIntensity+1))*(mxIntensity+1))
```

```python
bit_levels = [7, 6, 5, 4, 3, 2, 1, 0]
for bit in bit_levels:
    intensity_res = Decrease_Intensity_Resolution(gray_image, bit)
    plt.imshow(intensity_res, cmap='gray')
    plt.title(f'{bit}-bit intensity level')
    plt.show()
```
