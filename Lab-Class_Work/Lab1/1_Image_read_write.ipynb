```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Read the image using matplotlib's imread function
image = mpimg.imread('triangle.png')

# Display the image using imshow
plt.imshow(image)

# Optionally, turn off the axes for a cleaner look
plt.axis('off')

# Show the image
plt.show()
```

```python
# Convert RGB/BGR image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print('shape of the image: '+str(gray_image.shape))
```

```python
resized_image = cv2.resize(gray_image, (10, 10))
plt.imshow(resized_image,cmap='gray')
plt.grid(True)
print(resized_image.shape)
```

```python
print(resized_image[4:8,4:8])
```

```python
#histrogram plotting
import numpy as np
def imghist(img):
    hist = np.zeros(256)
    for i in range(256):
        hist[i] = sum(sum(img == i))
    return hist    
        
```

```python
resized_image_hist = imghist(resized_image)
print(resized_image_hist.shape)
plt.bar(np.arange(256),resized_image_hist)
plt.xlim(0,10)
plt.title('Histogram')
```

```python
# Plot the histogram of the grayscale image
import numpy as np
hist,bins = np.histogram(resized_image,bins=256,range=(0,2))
print(bins.shape)
plt.plot(bins[:-1],hist)
plt.show()
```
